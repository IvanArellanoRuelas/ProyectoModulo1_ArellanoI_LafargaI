### 1.1 Valuación de Créditos.

En este proyecto analizaremos varias variables que influyen en la decisión de asignación de crédito y categoría de tarjeta, que se le debe ofrecer al cliente para que el  se sienta cómodo con su tarjeta 


### 1.2 Objetivos.
